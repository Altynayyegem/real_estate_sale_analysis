```python
#This environment (VS code) unfortunately does not allow to create stored procedures, but if running below command in My SQL it will work
DELIMITER @
CREATE PROCEDURE UPDATE_SALES (IN pTown VARCHAR(20), IN pID INTEGER)

BEGIN
    IF pTown = 'Greenwich' THEN
        UPDATE Real_Estates_Sales
        SET Sale_Amount (USD) = Sale_Amount (USD) - (Sale_Amount (USD) * 0.25)
        WHERE Serial_Number = pID;
    ELSEIF pTown = 'Hamden' THEN
        UPDATE Real_Estates_Sales
        SET Sale_Amount (USD) = Sale_Amount (USD) - (Sale_Amount (USD) * 0.5)
        WHERE Serial_Number = pID;
    ELSE 
        UPDATE Real_Estates_Sales
        SET Sale_Amount (USD) = Sale_Amount (USD)
        WHERE Serial_Number = pID;
    END IF;

END @
DELIMITER ;
```
